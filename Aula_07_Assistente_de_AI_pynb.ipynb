```python
def somar_valores():
    total = 0
    # Exibe a instrução para o usuário inserir números
    print("Digite os números para somar (digite 'fim' para terminar):")
    while True:
        entrada = input("Número: ")
        # Verifica se o usuário deseja encerrar a entrada de números
        if entrada.lower() == 'fim':
            break
        try:
            numero = float(entrada)
            # Adiciona o número ao total acumulado
            total += numero
        except ValueError:
            # Trata entradas não numéricas
            print("Entrada inválida. Por favor, digite um número ou 'fim'.")
    # Retorna a soma total dos números
    return total

def subtrair_valores():
    primeiro_numero = None
    # Exibe a instrução para o usuário inserir números
    print("Digite os números para subtrair (digite 'fim' para terminar):")
    while True:
        entrada = input("Número: ")
        # Verifica se o usuário deseja encerrar a entrada de números
        if entrada.lower() == 'fim':
            break
        try:
            numero = float(entrada)
            # Se for o primeiro número, ele inicia a subtração
            if primeiro_numero is None:
                primeiro_numero = numero
                resultado = numero
            else:
                # Subtrai os números subsequentes do resultado
                resultado -= numero
        except ValueError:
            # Trata entradas não numéricas
            print("Entrada inválida. Por favor, digite um número ou 'fim'.")
    # Retorna o resultado da subtração ou 0 se nenhum número foi inserido
    if primeiro_numero is not None:
        return resultado
    else:
        return 0 # Caso nenhum número seja inserido

# Loop principal da calculadora
while True:
    # Exibe o menu de opções para o usuário
    print("\n--- Calculadora Simples ---")
    print("1. Somar")
    print("2. Subtrair")
    print("3. Sair")

    # Solicita a escolha do usuário
    escolha = input("Escolha uma operação (1/2/3): ")

    # Verifica a escolha do usuário e executa a operação correspondente
    if escolha == '1':
        # Chama a função de soma e exibe o resultado
        resultado = somar_valores()
        print(f"O resultado da soma é: {resultado}")
    elif escolha == '2':
        # Chama a função de subtração e exibe o resultado
        resultado = subtrair_valores()
        print(f"O resultado da subtração é: {resultado}")
    elif escolha == '3':
        # Encerra o programa
        print("Saindo da calculadora. Até mais!")
        break
    else:
        # Lida com opções inválidas
        print("Opção inválida. Por favor, escolha 1, 2 ou 3.")
```
