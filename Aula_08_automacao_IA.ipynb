```python
import random

# Simula a leitura da temperatura do sistema (em um cenário real, usaria um sensor ou API)
system_temperature = random.uniform(20.0, 35.0) # Gera um valor aleatório entre 20.0 e 35.0 para a temperatura

print(f"Temperatura atual do sistema: {system_temperature:.2f}°C")

# Avalia se a temperatura excede 30 graus Celsius
if system_temperature > 30:
    print("ALERTA: Superaquecimento detectado! Temperatura acima de 30°C.")
else:
    print("Temperatura do sistema está normal.")
```

```python
import os
import shutil

# --- CONFIGURAÇÃO --- #
# Define o diretório a ser organizado.
source_dir = './arquivos_para_organizar' # Altere conforme necessário

# Cria o diretório de origem se não existir (útil para testes).
if not os.path.exists(source_dir):
    os.makedirs(source_dir)
    print(f"Diretório de origem '{source_dir}' criado para demonstração. Adicione arquivos para testar.")

print(f"Organizando arquivos em: {os.path.abspath(source_dir)}")

# Itera sobre todos os itens no diretório de origem.
for filename in os.listdir(source_dir):
    file_path = os.path.join(source_dir, filename)

    # Processa apenas arquivos.
    if os.path.isfile(file_path):
        # Separa nome e extensão do arquivo.
        name, extension = os.path.splitext(filename)

        # Define o nome da pasta de destino com base na extensão.
        # Ex: .pdf -> PDFs, .jpg -> JPGs. Arquivos sem extensão vão para 'SEM_EXTENSAO'.
        if extension:
            folder_name = extension[1:].upper() + 's'
        else:
            folder_name = 'SEM_EXTENSAO'

        destination_folder = os.path.join(source_dir, folder_name)

        # Cria a pasta de destino se não existir.
        if not os.path.exists(destination_folder):
            os.makedirs(destination_folder)
            print(f"Pasta '{destination_folder}' criada.")

        # Move o arquivo para a pasta de destino.
        shutil.move(file_path, os.path.join(destination_folder, filename))
        print(f"Movido '{filename}' para '{destination_folder}'")

print("Organização de arquivos concluída!")
```

```python
import requests

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço de um determinado CEP.

    Args:
        cep (str): O Código de Endereçamento Postal (CEP) a ser consultado.

    Returns:
        dict or None: Um dicionário com os dados do endereço formatados se a consulta for bem-sucedida,
                      ou None em caso de erro ou CEP não encontrado.
    """
    # Remove caracteres não numéricos do CEP.
    cep = ''.join(filter(str.isdigit, cep))

    # Verifica se o CEP tem 8 dígitos.
    if len(cep) != 8:
        print("Erro: O CEP deve conter 8 dígitos numéricos.")
        return None

    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro para códigos de status HTTP ruins (4xx ou 5xx).

        data = response.json()

        # Verifica se a API retornou um erro (ex: CEP não encontrado).
        if 'erro' in data and data['erro']:
            print(f"Erro: CEP '{cep}' não encontrado.")
            return None

        # Formata as informações para exibição legível.
        endereco_formatado = {
            "CEP": data.get('cep'),
            "Logradouro": data.get('logradouro'),
            "Complemento": data.get('complemento'),
            "Bairro": data.get('bairro'),
            "Cidade": data.get('localidade'),
            "Estado": data.get('uf'),
            "IBGE": data.get('ibge'),
            "GIA": data.get('gia'),
            "DDD": data.get('ddd'),
            "SIAFI": data.get('siafi')
        }
        return endereco_formatado

    except requests.exceptions.RequestException as e:
        print(f"Erro de conexão ou requisição: {e}")
        return None
    except ValueError:
        print("Erro: Não foi possível decodificar a resposta JSON da API.")
        return None

# --- Uso interativo ---
cep_digitado = input("Por favor, digite o CEP que deseja consultar: ")
print(f"\nConsultando CEP: {cep_digitado}")
endereco = consultar_cep(cep_digitado)

if endereco:
    print("\n--- Informações do Endereço ---")
    for chave, valor in endereco.items():
        if valor: # Exibe apenas os campos que possuem valor.
            print(f"{chave}: {valor}")
else:
    print(f"Não foi possível obter informações para o CEP '{cep_digitado}'.")
```

```python
import random
import time

# --- Configurações do Monitoramento ---
CRITICAL_TEMP_THRESHOLD = 30.0 # Temperatura em °C que dispara o alerta
MONITOR_INTERVAL_SECONDS = 5   # Intervalo de verificação em segundos

def simulate_system_temperature():
    """
    Simula a leitura da temperatura atual do sistema.
    Em um cenário real, usaria um sensor ou API.
    """
    return random.uniform(25.0, 38.0) # Gera temperaturas variadas

def send_alert(message):
    """
    Dispara um alerta, imprimindo-o no console.
    Pode ser estendido para e-mail, SMS, etc.
    """
    print(f"\n{'='*30}\nALERTA CRÍTICO: {message}\n{'='*30}")
    # Integração com serviço de e-mail/SMS pode ser adicionada aqui.

def autonomous_monitor():
    print("Iniciando monitoramento autônomo de temperatura...")
    print(f"Limite crítico: {CRITICAL_TEMP_THRESHOLD}°C. Verificando a cada {MONITOR_INTERVAL_SECONDS} segundos.")

    alert_triggered = False # Flag para evitar alertas repetidos na mesma condição

    try:
        while True:
            current_temperature = simulate_system_temperature()
            print(f"[{(time.strftime('%H:%M:%S'))}] Temperatura atual: {current_temperature:.2f}°C", end='\r')

            if current_temperature > CRITICAL_TEMP_THRESHOLD:
                if not alert_triggered:
                    send_alert(f"Superaquecimento detectado! Temp: {current_temperature:.2f}°C (acima de {CRITICAL_TEMP_THRESHOLD}°C)")
                    alert_triggered = True
            else:
                if alert_triggered:
                    print(f"[{(time.strftime('%H:%M:%S'))}] Temperatura voltou ao normal: {current_temperature:.2f}°C")
                alert_triggered = False

            time.sleep(MONITOR_INTERVAL_SECONDS)

    except KeyboardInterrupt:
        print("\nMonitoramento interrompido pelo usuário.")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")

# Inicia o programa autônomo
autonomous_monitor()
```
